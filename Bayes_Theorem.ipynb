```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import seaborn as sns

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("="*60)
print("    BAYESIAN MEDICAL TUMOR DETECTION SYSTEM")
print("="*60)
print()

class BayesianTumorDetector:
    """
    A Bayesian system for detecting tumors in medical scan images.

    Real-world scenario: A hospital's radiology department uses AI-assisted
    diagnosis to help radiologists identify potential tumors in CT scans,
    MRI scans, and X-rays. The system combines multiple imaging indicators
    with patient risk factors using Bayesian inference.
    """

    def __init__(self):
        self.history = []
        self.prior_tumor_probability = 0.12  # Initial belief: 12% of scans show tumors (based on hospital statistics)

    def bayes_theorem(self, prior, likelihood_evidence_given_tumor, likelihood_evidence_given_no_tumor):
        """
        Apply Bayes' theorem to calculate posterior probability

        P(Tumor|Evidence) = P(Evidence|Tumor) * P(Tumor) / P(Evidence)

        where P(Evidence) = P(Evidence|Tumor)*P(Tumor) + P(Evidence|No Tumor)*P(No Tumor)
        """
        # Calculate marginal probability (total probability of evidence)
        marginal_probability = (likelihood_evidence_given_tumor * prior +
                              likelihood_evidence_given_no_tumor * (1 - prior))

        # Calculate posterior probability
        posterior = (likelihood_evidence_given_tumor * prior) / marginal_probability

        return posterior, marginal_probability

    def update_belief(self, evidence_type, evidence_value, scenario_name):
        """
        Update our belief about tumor probability based on new evidence
        """
        print(f"\n{'='*60}")
        print(f"MEDICAL CASE: {scenario_name}")
        print(f"{'='*60}")

        # Define likelihood probabilities based on evidence type
        if evidence_type == "suspicious_mass":
            # Suspicious mass detected by image analysis
            likelihood_tumor = 0.85      # P(Suspicious Mass | Tumor)
            likelihood_no_tumor = 0.08  # P(Suspicious Mass | No Tumor)

        elif evidence_type == "irregular_shape":
            # Irregular/spiculated edges often indicate malignancy
            likelihood_tumor = 0.78      # P(Irregular Shape | Tumor)
            likelihood_no_tumor = 0.15  # P(Irregular Shape | No Tumor)

        elif evidence_type == "high_contrast_enhancement":
            # Contrast agent uptake patterns
            likelihood_tumor = 0.72      # P(High Enhancement | Tumor)
            likelihood_no_tumor = 0.25  # P(High Enhancement | No Tumor)

        elif evidence_type == "family_history":
            # Genetic predisposition increases risk
            likelihood_tumor = 0.65      # P(Family History | Tumor)
            likelihood_no_tumor = 0.20  # P(Family History | No Tumor)

        elif evidence_type == "abnormal_texture":
            # Texture analysis shows heterogeneous patterns
            likelihood_tumor = 0.80      # P(Abnormal Texture | Tumor)
            likelihood_no_tumor = 0.12  # P(Abnormal Texture | No Tumor)

        elif evidence_type == "elevated_biomarkers":
            # Blood test shows elevated tumor markers
            likelihood_tumor = 0.75      # P(Elevated Markers | Tumor)
            likelihood_no_tumor = 0.18  # P(Elevated Markers | No Tumor)

        elif evidence_type == "size_increase":
            # Comparison with previous scan shows growth
            likelihood_tumor = 0.88      # P(Size Increase | Tumor)
            likelihood_no_tumor = 0.05  # P(Size Increase | No Tumor)

        # Current prior (from previous update or initial)
        current_prior = self.prior_tumor_probability

        print(f"üî¨ Evidence Type: {evidence_type.replace('_', ' ').title()}")
        print(f"üìä Evidence Details: {evidence_value}")
        print(f"üë§ Patient ID: {scenario_name.split(' - ')[0]}")
        print()
        print("STEP 1: Current Probabilities (Prior Belief)")
        print(f"Prior P(Tumor) = {current_prior:.4f} ({current_prior*100:.1f}%)")
        print(f"Prior P(No Tumor) = {1 - current_prior:.4f} ({(1-current_prior)*100:.1f}%)")
        print()
        print("STEP 2: Likelihood Probabilities (Evidence Strength)")
        print(f"P(Evidence | Tumor Present) = {likelihood_tumor:.4f}")
        print(f"P(Evidence | No Tumor) = {likelihood_no_tumor:.4f}")
        print(f"Likelihood Ratio = {likelihood_tumor/likelihood_no_tumor:.2f} (Evidence is {likelihood_tumor/likelihood_no_tumor:.1f}x more likely if tumor present)")
        print()

        # Apply Bayes' theorem
        posterior, marginal = self.bayes_theorem(current_prior, likelihood_tumor, likelihood_no_tumor)

        print("STEP 3: Bayes' Theorem Calculation")
        print("P(Tumor | Evidence) = P(Evidence | Tumor) √ó P(Tumor) / P(Evidence)")
        print()
        print("First, calculate P(Evidence):")
        print(f"P(Evidence) = P(Evidence|Tumor)√óP(Tumor) + P(Evidence|No Tumor)√óP(No Tumor)")
        print(f"P(Evidence) = {likelihood_tumor:.4f}√ó{current_prior:.4f} + {likelihood_no_tumor:.4f}√ó{1-current_prior:.4f}")
        print(f"P(Evidence) = {likelihood_tumor * current_prior:.4f} + {likelihood_no_tumor * (1-current_prior):.4f}")
        print(f"P(Evidence) = {marginal:.4f}")
        print()
        print("Now calculate posterior:")
        print(f"P(Tumor | Evidence) = ({likelihood_tumor:.4f} √ó {current_prior:.4f}) / {marginal:.4f}")
        print(f"P(Tumor | Evidence) = {likelihood_tumor * current_prior:.4f} / {marginal:.4f}")
        print(f"P(Tumor | Evidence) = {posterior:.4f}")
        print()

        # Store history for visualization
        self.history.append({
            'scenario': scenario_name,
            'evidence_type': evidence_type,
            'prior': current_prior,
            'likelihood_tumor': likelihood_tumor,
            'likelihood_no_tumor': likelihood_no_tumor,
            'marginal': marginal,
            'posterior': posterior,
            'evidence_value': evidence_value
        })

        # Update prior for next iteration
        self.prior_tumor_probability = posterior

        print("STEP 4: Updated Medical Assessment")
        print(f"üéØ New P(Tumor) = {posterior:.4f} ({posterior*100:.1f}%)")
        print(f"üéØ New P(No Tumor) = {1-posterior:.4f} ({(1-posterior)*100:.1f}%)")

        # Medical decision thresholds
        high_risk_threshold = 0.70  # Immediate intervention
        moderate_risk_threshold = 0.30  # Additional testing

        if posterior > high_risk_threshold:
            decision = "üö® HIGH RISK - IMMEDIATE BIOPSY/TREATMENT RECOMMENDED"
            risk_level = "HIGH"
        elif posterior > moderate_risk_threshold:
            decision = "‚ö†Ô∏è  MODERATE RISK - ADDITIONAL TESTING REQUIRED"
            risk_level = "MODERATE"
        else:
            decision = "‚úÖ LOW RISK - ROUTINE FOLLOW-UP"
            risk_level = "LOW"

        print(f"\nüè• MEDICAL DECISION: {decision}")
        print(f"üìã Risk Level: {risk_level}")
        print(f"üìà Confidence Change: {((posterior/current_prior - 1) * 100):+.1f}%")

        return posterior

# Initialize the Bayesian tumor detector
detector = BayesianTumorDetector()

print("üè• MEDICAL SCENARIO:")
print("A 58-year-old patient undergoes routine screening. The radiology")
print("department uses a Bayesian AI system to assist in tumor detection")
print("by analyzing multiple imaging features and patient risk factors.")
print("Each piece of evidence updates the probability assessment.")
print()
print(f"üìä Hospital Base Rate: {detector.prior_tumor_probability:.1%} of scans show tumors")
print(f"üë§ Patient: Case #A12847 - Initial Risk Assessment")

# Medical Case: Sequential evidence analysis
print("\n" + "="*80)
print("                    SEQUENTIAL DIAGNOSTIC PROCESS")
print("="*80)

# Evidence 1: AI detects suspicious mass
posterior_1 = detector.update_belief(
    "suspicious_mass",
    "AI algorithm detected 2.3cm irregular mass in upper lobe",
    "Patient A12847 - CT Scan Analysis"
)

# Evidence 2: Radiologist confirms irregular shape
posterior_2 = detector.update_belief(
    "irregular_shape",
    "Spiculated margins with 85% confidence from shape analysis",
    "Patient A12847 - Morphological Assessment"
)

# Evidence 3: Contrast enhancement analysis
posterior_3 = detector.update_belief(
    "high_contrast_enhancement",
    "Contrast uptake 3.2x higher than surrounding tissue",
    "Patient A12847 - Contrast Study Results"
)

# Evidence 4: Patient history review
posterior_4 = detector.update_belief(
    "family_history",
    "Family history: Mother had lung cancer at age 62",
    "Patient A12847 - Medical History Review"
)

# Evidence 5: Texture analysis
posterior_5 = detector.update_belief(
    "abnormal_texture",
    "Heterogeneous texture pattern with 78% abnormality score",
    "Patient A12847 - Advanced Texture Analysis"
)

# Evidence 6: Biomarker results
posterior_6 = detector.update_belief(
    "elevated_biomarkers",
    "CEA: 8.2 ng/mL (normal <3.0), CYFRA 21-1: 4.1 ng/mL (normal <3.3)",
    "Patient A12847 - Laboratory Results"
)

# Evidence 7: Comparison with previous scan
posterior_7 = detector.update_belief(
    "size_increase",
    "25% size increase compared to scan from 6 months ago",
    "Patient A12847 - Temporal Comparison"
)

# Create comprehensive medical visualizations
fig = plt.figure(figsize=(20, 16))

# Plot 1: Diagnostic probability evolution
ax1 = plt.subplot(3, 3, 1)
scenarios = ["Initial\nScreening"] + [h['evidence_type'].replace('_', ' ').title() for h in detector.history]
probabilities = [0.12] + [h['posterior'] for h in detector.history]

# Color code based on risk levels
colors = []
for p in probabilities:
    if p > 0.70:
        colors.append('red')      # High risk
    elif p > 0.30:
        colors.append('orange')   # Moderate risk
    else:
        colors.append('green')    # Low risk

bars = ax1.bar(range(len(scenarios)), probabilities, color=colors, alpha=0.7)
ax1.set_title('üéØ Tumor Probability Evolution', fontsize=14, fontweight='bold')
ax1.set_ylabel('P(Tumor | Evidence)')
ax1.set_ylim(0, 1)
ax1.axhline(y=0.70, color='red', linestyle='--', alpha=0.7, label='High Risk (70%)')
ax1.axhline(y=0.30, color='orange', linestyle='--', alpha=0.7, label='Moderate Risk (30%)')

# Add percentage labels on bars
for bar, prob in zip(bars, probabilities):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{prob:.1%}', ha='center', va='bottom', fontweight='bold')

ax1.set_xticks(range(len(scenarios)))
ax1.set_xticklabels(scenarios, rotation=45, ha='right')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Likelihood strength analysis
ax2 = plt.subplot(3, 3, 2)
evidence_types = [h['evidence_type'].replace('_', ' ').title() for h in detector.history]
likelihood_ratios = [h['likelihood_tumor'] / h['likelihood_no_tumor'] for h in detector.history]

# Color by strength
colors = ['darkred' if lr > 10 else 'red' if lr > 5 else 'orange' if lr > 2 else 'yellow' for lr in likelihood_ratios]
bars = ax2.barh(evidence_types, likelihood_ratios, color=colors, alpha=0.7)
ax2.set_title('üî¨ Evidence Strength (Likelihood Ratios)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Likelihood Ratio (P(Evidence|Tumor) / P(Evidence|No Tumor))')
ax2.axvline(x=1, color='black', linestyle='--', alpha=0.5, label='Neutral Evidence')
ax2.set_xscale('log')
ax2.grid(True, alpha=0.3)

# Add ratio labels
for bar, lr in zip(bars, likelihood_ratios):
    width = bar.get_width()
    ax2.text(width + 0.1, bar.get_y() + bar.get_height()/2.,
             f'{lr:.1f}x', ha='left', va='center', fontweight='bold')

# Plot 3: Diagnostic confidence over time
ax3 = plt.subplot(3, 3, 3)
confidence_scores = [abs(h['posterior'] - 0.5) * 2 for h in detector.history]
test_numbers = range(1, len(confidence_scores) + 1)

ax3.plot(test_numbers, confidence_scores, 'o-', linewidth=3, markersize=8, color='purple')
ax3.fill_between(test_numbers, confidence_scores, alpha=0.3, color='purple')
ax3.set_title('üìà Diagnostic Confidence Progression', fontsize=14, fontweight='bold')
ax3.set_xlabel('Test Number')
ax3.set_ylabel('Diagnostic Confidence')
ax3.set_ylim(0, 1)
ax3.grid(True, alpha=0.3)

# Plot 4: Prior vs Posterior comparison
ax4 = plt.subplot(3, 3, 4)
priors = [h['prior'] for h in detector.history]
posteriors = [h['posterior'] for h in detector.history]

x = np.arange(len(evidence_types))
width = 0.35

bars1 = ax4.bar(x - width/2, priors, width, label='Prior P(Tumor)', alpha=0.8, color='lightblue')
bars2 = ax4.bar(x + width/2, posteriors, width, label='Posterior P(Tumor)', alpha=0.8, color='darkblue')

ax4.set_title('üìä Before vs After Each Test', fontsize=14, fontweight='bold')
ax4.set_ylabel('Probability')
ax4.set_xticks(x)
ax4.set_xticklabels([et.replace(' ', '\n') for et in evidence_types], rotation=0, fontsize=9)
ax4.legend()
ax4.grid(True, alpha=0.3)

# Plot 5: Cumulative evidence impact
ax5 = plt.subplot(3, 3, 5)
cumulative_odds = []
current_odds = 0.12 / (1 - 0.12)  # Convert initial prior to odds

odds_history = [current_odds]
for h in detector.history:
    bayes_factor = h['likelihood_tumor'] / h['likelihood_no_tumor']
    current_odds *= bayes_factor
    odds_history.append(current_odds)

ax5.plot(range(len(odds_history)), odds_history, 's-', linewidth=2, markersize=6, color='red')
ax5.set_title('‚öñÔ∏è Cumulative Evidence (Odds Ratio)', fontsize=14, fontweight='bold')
ax5.set_xlabel('Evidence Count')
ax5.set_ylabel('Odds Ratio (Tumor:No Tumor)')
ax5.set_yscale('log')
ax5.grid(True, alpha=0.3)

# Add odds labels
for i, odds in enumerate(odds_history[::2]):  # Show every other point to avoid clutter
    ax5.annotate(f'{odds:.1f}:1', (i*2, odds), textcoords="offset points",
                xytext=(0,10), ha='center', fontsize=8)

# Plot 6: Risk category evolution
ax6 = plt.subplot(3, 3, 6)
risk_categories = []
for p in [0.12] + [h['posterior'] for h in detector.history]:
    if p > 0.70:
        risk_categories.append('High Risk')
    elif p > 0.30:
        risk_categories.append('Moderate Risk')
    else:
        risk_categories.append('Low Risk')

# Count risk levels over time
risk_colors = {'Low Risk': 'green', 'Moderate Risk': 'orange', 'High Risk': 'red'}
unique_risks = list(set(risk_categories))
risk_counts = {risk: [1 if cat == risk else 0 for cat in risk_categories] for risk in unique_risks}

bottom = np.zeros(len(risk_categories))
for risk in unique_risks:
    ax6.bar(range(len(risk_categories)), risk_counts[risk], bottom=bottom,
            label=risk, color=risk_colors[risk], alpha=0.7)
    bottom += risk_counts[risk]

ax6.set_title('üö¶ Risk Category Progression', fontsize=14, fontweight='bold')
ax6.set_xlabel('Assessment Stage')
ax6.set_ylabel('Risk Level')
ax6.legend()
ax6.set_xticks(range(len(risk_categories)))
ax6.set_xticklabels(['Initial'] + [f'Test {i+1}' for i in range(len(detector.history))], rotation=45)

# Plot 7: Sensitivity analysis heatmap
ax7 = plt.subplot(3, 3, 7)
sensitivity_data = np.array([[h['likelihood_tumor'], h['likelihood_no_tumor']] for h in detector.history])
im = ax7.imshow(sensitivity_data.T, cmap='RdYlBu_r', aspect='auto', vmin=0, vmax=1)
ax7.set_title('üé® Evidence Likelihood Heatmap', fontsize=14, fontweight='bold')
ax7.set_xlabel('Evidence Type')
ax7.set_ylabel('Condition')
ax7.set_xticks(range(len(evidence_types)))
ax7.set_xticklabels([et.replace(' ', '\n') for et in evidence_types], rotation=0, fontsize=8)
ax7.set_yticks([0, 1])
ax7.set_yticklabels(['P(Evidence|Tumor)', 'P(Evidence|No Tumor)'])
plt.colorbar(im, ax=ax7)

# Plot 8: Bayesian update magnitude
ax8 = plt.subplot(3, 3, 8)
update_magnitudes = []
prev_prob = 0.12
for h in detector.history:
    magnitude = abs(h['posterior'] - prev_prob)
    update_magnitudes.append(magnitude)
    prev_prob = h['posterior']

bars = ax8.bar(evidence_types, update_magnitudes, color='purple', alpha=0.6)
ax8.set_title('üìà Bayesian Update Magnitude', fontsize=14, fontweight='bold')
ax8.set_ylabel('|Posterior - Prior|')
ax8.set_xticklabels(evidence_types, rotation=45, ha='right')
ax8.grid(True, alpha=0.3)

# Plot 9: Final diagnostic summary
ax9 = plt.subplot(3, 3, 9)
final_prob = detector.prior_tumor_probability
ax9.pie([final_prob, 1-final_prob],
        labels=[f'Tumor\n{final_prob:.1%}', f'No Tumor\n{(1-final_prob):.1%}'],
        colors=['red', 'lightgreen'], autopct='%1.1f%%', startangle=90)
ax9.set_title('üéØ Final Diagnostic Assessment', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Medical summary report
print("\n" + "="*80)
print("                    MEDICAL DIAGNOSTIC REPORT")
print("="*80)

summary_df = pd.DataFrame([
    {
        'Test Order': i+1,
        'Evidence Type': h['evidence_type'].replace('_', ' ').title(),
        'Prior P(Tumor)': f"{h['prior']:.3f}",
        'Likelihood Ratio': f"{h['likelihood_tumor']/h['likelihood_no_tumor']:.1f}",
        'Posterior P(Tumor)': f"{h['posterior']:.3f}",
        'Risk Level': 'HIGH' if h['posterior'] > 0.7 else 'MODERATE' if h['posterior'] > 0.3 else 'LOW'
    }
    for i, h in enumerate(detector.history)
])

print(summary_df.to_string(index=False))

# Clinical significance explanation
print("\n" + "="*80)
print("                    CLINICAL SIGNIFICANCE & IMPACT")
print("="*80)

print(f"""
üè• PATIENT CASE SUMMARY:
   Patient A12847 underwent comprehensive diagnostic workup using Bayesian inference

üìä DIAGNOSTIC PROGRESSION:
   ‚Ä¢ Initial screening probability: 12.0%
   ‚Ä¢ Final diagnostic probability: {detector.prior_tumor_probability:.1f}%
   ‚Ä¢ Total evidence pieces analyzed: {len(detector.history)}

üéØ CLINICAL DECISION SUPPORT:

1. EVIDENCE-BASED MEDICINE:
   ‚Ä¢ Each diagnostic test provides quantified evidence
   ‚Ä¢ Combines multiple imaging and clinical indicators
   ‚Ä¢ Updates probability assessment in real-time
   ‚Ä¢ Reduces diagnostic uncertainty through systematic analysis

2. PERSONALIZED RISK ASSESSMENT:
   ‚Ä¢ Incorporates patient-specific factors (family history)
   ‚Ä¢ Weighs imaging findings with clinical context
   ‚Ä¢ Provides individualized probability estimates
   ‚Ä¢ Supports personalized treatment planning

3. DIAGNOSTIC ACCURACY IMPROVEMENT:
   ‚Ä¢ Reduces false positives through systematic evaluation
   ‚Ä¢ Minimizes false negatives with comprehensive analysis
   ‚Ä¢ Quantifies diagnostic confidence levels
   ‚Ä¢ Supports radiologist decision-making

4. HEALTHCARE RESOURCE OPTIMIZATION:
   ‚Ä¢ Prioritizes high-risk cases for immediate attention
   ‚Ä¢ Guides appropriate follow-up intervals
   ‚Ä¢ Optimizes biopsy and treatment decisions
   ‚Ä¢ Reduces unnecessary procedures for low-risk cases

5. INTERPRETABLE AI IN MEDICINE:
   ‚Ä¢ Every probability calculation is mathematically transparent
   ‚Ä¢ Clinical reasoning process is auditable and explainable
   ‚Ä¢ Supports medical education and training
   ‚Ä¢ Enables quality assurance and peer review

üö® FINAL CLINICAL RECOMMENDATION:
   Based on Bayesian analysis: {detector.prior_tumor_probability:.1f}% probability of malignancy

   {"üö® IMMEDIATE ACTION REQUIRED:" if detector.prior_tumor_probability > 0.7 else
    "‚ö†Ô∏è  ADDITIONAL MONITORING RECOMMENDED:" if detector.prior_tumor_probability > 0.3 else
    "‚úÖ ROUTINE FOLLOW-UP APPROPRIATE:"}

   {'''‚Ä¢ Urgent referral to oncology
   ‚Ä¢ Tissue biopsy within 48 hours
   ‚Ä¢ Staging studies if confirmed
   ‚Ä¢ Patient counseling and support''' if detector.prior_tumor_probability > 0.7 else
    '''‚Ä¢ Additional imaging in 3 months
   ‚Ä¢ Consider PET scan for staging
   ‚Ä¢ Multidisciplinary team review
   ‚Ä¢ Close monitoring protocol''' if detector.prior_tumor_probability > 0.3 else
    '''‚Ä¢ Routine follow-up in 12 months
   ‚Ä¢ Continue regular screening
   ‚Ä¢ Patient reassurance appropriate
   ‚Ä¢ Standard surveillance protocol'''}
""")

print("\nüßÆ BAYES' THEOREM IN MEDICAL DIAGNOSIS:")
print("P(Disease|Test) = P(Test|Disease) √ó P(Disease) / P(Test)")
print("\nThis fundamental formula enables:")
print("‚Ä¢ Rational updating of diagnostic probabilities")
print("‚Ä¢ Integration of multiple diagnostic tests")
print("‚Ä¢ Quantification of diagnostic uncertainty")
print("‚Ä¢ Evidence-based clinical decision making")

print(f"\nüí° KEY INSIGHT:")
print(f"The patient's tumor probability increased from 12% to {detector.prior_tumor_probability:.1f}%")
print(f"through systematic Bayesian evidence accumulation - a {((detector.prior_tumor_probability/0.12 - 1) * 100):.0f}% increase")
print("demonstrating the power of combining multiple diagnostic indicators.")
```
